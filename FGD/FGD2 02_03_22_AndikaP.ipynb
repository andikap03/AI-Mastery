```python
from astropy.table import Table

print("Menu Jajanan :")
data_rows = [(1, "Donat", "Rp 1250"),
             (2, "Bakwan", "Rp 1000"),
             (3, "Onde-onde", "Rp 800"),
             (4, "Lemper", "Rp 1250"),
             (5, "Risol", "Rp 1500")]
t = Table(rows=data_rows, names=('No', 'Menu','Harga'))
print(t)
print()

menu = {'Donat' : 1250,
        'Bakwan' : 1000,
        'Onde-onde' : 800,
        'Lemper' : 1250,
        'Risol' : 1500}
makanan = ['Donat','Bakwan','Onde-onde','Lemper','Risol']
pesanan = input('Pesan menu nomer :').lower()
jumlah = int(input('Jumlahnya :'))

if pesanan == 'cukup':
  print('Terima kasih sudah berbelanja jajanan pasar kami!')
else:
  num = [str(i) for i in range(1, len(makanan)+1)]
  if pesanan in num:
    if int(pesanan) <= len(makanan):
      for pesan in range(len(makanan)):    
        if int(pesanan) == pesan+1:
          print('Kamu memesan', jumlah, 'buah', makanan[int(pesanan)-1], ', total harga Rp', menu[makanan[int(pesanan)-1]] * jumlah, ', terimakasih sudah berbelanja!')
  else:
    print("Menu hanya ada", len(makanan), ", dan isikan dengan nomor!")
```

```python
from astropy.table import Table

print("List Merk Mobil :")
data_rows = [(1, "Ford", "$ 100"),
             (2, "Mercedes", "$ 125"),
             (3, "Tesla", "$ 267"),
             (4, "Honda", "$ 78"),
             (5, "Suzuki", "$ 90")]
t = Table(rows=data_rows, names=('No', 'Merk','Harga'))
print(t)
print()

mobil = {'Ford' : 100,
        'Mercedes' : 125,
        'Tesla' : 267,
        'Honda' : 78,
        'Suzuki' : 90}
merk = ['Ford','Mercedes','Tesla','Honda','Suzuki']
konversi = 14000

while True:
  pesanan = input('Pesan Merk Nomer :').lower()
  if pesanan == 'cancel':
    print('Terima kasih, selamat datang kembali !')
    break
  else:
    num = [str(i) for i in range(1, len(merk)+1)]
    if pesanan in num:
      for pesan in range(len(merk)):    
        if int(pesanan) == pesan+1:
          print('Kamu memesan mobil merk', merk[int(pesanan)-1], '. Total harga $', mobil[merk[int(pesanan)-1]], 'atau sebesar Rp' , mobil[merk[int(pesanan)-1]] * konversi,', selamat datang kembali !')
      break  
    else:
      print("Merk yang tersedia hanya ada", len(merk), ". Harap isikan dengan nomor!")
```
