```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
table_data=[
    ["Player 1", "Player 2", "Pemenang"],
    ["Batu", "Gunting", "Player 1"],
    ["Gunting", "Batu", "Player 2"],
    ["Batu", "Kertas", "Player 2"],
    ["Kertas", "Batu", "Player 1"],
    ["Gunting", "Kertas", "Player 1"],
    ["Kertas", "Gunting", "Player 2"],
    ["Kertas", "Kertas", "Seri"],
    ["Gunting", "Gunting", "Seri"],
    ["Batu", "Batu", "Seri"],
]
table = ax.table(cellText=table_data, loc='center')
table.set_fontsize(14)
table.scale(1,2)
ax.axis('off')
plt.show()
```

```python
Player_1 = input('Player 1 :').lower()
Player_2 = input('Player_2 :').lower()
valid = ['batu','gunting','kertas']

print('Pemenang : ')
if Player_1 in valid and Player_2 in valid:
  if Player_1 == 'batu' and Player_2 == 'kertas':
    print('Player 2')
  elif Player_1 == 'kertas' and Player_2 == 'batu':
    print('Player 1')
  elif Player_1 == 'gunting' and Player_2 == 'kertas':
    print('Player 1')
  elif Player_1 == 'kertas' and Player_2 == 'gunting':
    print('Player 2')
  elif Player_1 == 'batu' and Player_2 == 'gunting':
    print('Player 1')
  elif Player_1 == 'gunting' and Player_2 == 'batu':
    print('Player 2')
  else:
    print('Seri')
else:
  print('Input Tidak valid')
```

```python
from astropy.table import Table

data_rows = [("Donat", "Rp 1250"),
             ("Bakwan", "Rp 1000"),
             ("Onde-onde", "Rp 800"),
             ("Lemper", "Rp 1250"),
             ("Risol", "Rp 1500")]
t = Table(rows=data_rows, names=('Menu','Harga'))
print(t)
```

```python
menu = {'Donat' : 1250,
        'Bakwan' : 1000,
        'Onde-onde' : 800,
        'Lemper' : 1250,
        'Risol' : 1500}
pesan = input('Pesan menu nomer :')

if pesan == 'cukup':
  print('Terima kasih sudah berbelanja jajanan pasar kami!')
else:
  pesanan = int(pesan)
  if pesanan == 1:
    print('Kamu memesan Donat dengan harga Rp', menu['Donat'], ', terimakasih sudah berbelanja') 
  elif pesanan == 2:
    print('Kamu memesan Bakwan dengan harga Rp', menu['Bakwan'], ', terimakasih sudah berbelanja')
  elif pesanan == 3:
    print('Kamu memesan Onde-onde dengan harga Rp', menu['Onde-onde'], ', terimakasih sudah berbelanja')
  elif pesanan == 4:
    print('Kamu memesan Lemper dengan harga Rp', menu['Lemper'], ', terimakasih sudah berbelanja')  
  elif pesanan == 5:
    print('Kamu memesan Risol dengan harga Rp', menu['Risol'], ', terimakasih sudah berbelanja')  
  else:
    print('Menu Kosong!!')
```

```python
menu = {'Donat' : 1250,
        'Bakwan' : 1000,
        'Onde-onde' : 800,
        'Lemper' : 1250,
        'Risol' : 1500}
makanan = ['Donat','Bakwan','Onde-onde','Lemper','Risol']
pesanan = input('Pesan menu nomer :').lower()

if pesanan == 'cukup':
  print('Terima kasih sudah berbelanja jajanan pasar kami!')
else:
  num = [str(i) for i in range(len(makanan)+1)]
  if pesanan in num:
    if 0 < int(pesanan) <= len(makanan):
      for pesan in range(len(makanan)):    # [0,1,2,3,4] length = 5
        if int(pesanan) == pesan+1:
          print('Kamu memesan', makanan[int(pesanan)-1], 'dengan harga Rp', menu[makanan[int(pesanan)-1]], ', terimakasih sudah berbelanja!')
    else:
      print('Kosong! Menu hanya ada', len(makanan))
  else:
    print("Tak teridentifikasi, isikan dengan nomor!")
```
