## **A. Importing and Exporting JSON files**
The full form of JSON is JavaScript Object Notation. It means that a script (executable) file which is made of text in a programming language, is used to store and transfer the data. Python supports JSON through a built-in package called json.

The below code will write contents to the json file.

```python
import json
  
# Data to be written
dictionary ={"name" : "Meghan",
             "rollno" : 56,
             "cgpa" : 8.6,
             "age": 23}
  
# Serializing json 
json_object = json.dumps(dictionary, indent = 4)
  
# Writing to sample.json
with open("sample.json", "w") as outfile:
    outfile.write(json_object)
```

This will read contents from the json file and output it to the user.

```python
# Opening JSON file
with open('sample.json', 'r') as openfile:
  
    # Reading from json file
    json_object = json.load(openfile)
  
print(json_object)
print(type(json_object))
```

## **C. Analysing Outliers**
An outlier is a data point in a data set that is distant from all other observations. A data point that lies outside the overall distribution of the dataset.

Criteria to identify outliers:

* Data point that falls outside of 1.5 times of an interquartile range above the 3rd quartile and below the 1st quartile.

* Data point that falls outside of 3 standard deviations. we can use a z score and if the z score falls outside of 2 standard deviations

We have a data stating heights of certain individuals and we try to detect outliers.

```python
height_data = {'name':['Adit', 'Gia', 'Tim', 'Mark', 'Nur', 'Adil', 'Kia', 'Indar', 'Tia', 'Jack', 'Adele', 'Niall', 'Harry', 'Tom'],
               'height': [5.9, 5.2, 5.1, 5.5, 4.9, 5.4, 6.2, 6.5, 7.1, 14.5, 6.1, 5.6, 1.2, 5.5]}
df = pd.DataFrame(height_data)
df.T
```

Taking the concept of median a step further, we can define the Inter-Quartile Range (IQR). IQR is a measure of variability and is based on dividing a data set into quartiles. Quartile is the division of a set of observations into four intervals based on the values of the data.


### **1. Interquartile Range**
Basically, Quartiles divide a rank-ordered data set into four equal parts. The values that divide each part are called the first, second, and third quartiles where:

* Q1 is the "middle" value in the first half of the rank-ordered data set (25%). Also called lower-quartile

* Q2 is the median value in the set (50%).

* Q3 is the "middle" value in the second half of the rank-ordered data set (75%). Also called upper-quartile

The Inter-Quartile Range is the difference between the upper and lower quartiles:

>$IQR = Q3 - Q1 $ 

Where:

* Q1 = Median of First half of observations arranged in Ascending Order

* Q3 = Median of 2nd Half of observations arranged in Ascending Order

```python
# Find the upper bound and the lower bound
max_threshold = df['height'].quantile(0.95)
min_threshold = df['height'].quantile(0.05)

print('The upper bound :', max_threshold)
print('The upper bound :', min_threshold)
```

```python
# Find the outlier
upper_outlier = df[df['height'] > max_threshold]      #This is an outlier, above than the maximum limit
lower_outlier = df[df['height'] < min_threshold]      #This is an outlier, below the minimum limit
df_new = pd.concat([upper_outlier, lower_outlier])
df_new['status'] = ['Upper Outlier', 'Lower Outlier']
df_new
```

```python
# Remove Outlier
no_outlier = (df['height'] < max_threshold) & (df['height'] > min_threshold) #These are the values in between the two limits, i.e. without outliers
new_df = df[no_outlier] 
new_df.T 
```

## **A. Creating new categorical features from continuous variable**
Binning, also known as quantization is used for transforming continuous numeric features into discrete ones (categories). These discrete values or numbers can be thought of as categories or bins into which the raw, continuous numeric values are binned or grouped into.

Why do we need to create new categorical features from continous variables? Wide range of numerical data will be more readable if it is in groups statistical analysis of groups will provide better insight.

Let's say we want to group age by:

* 0 to 2 = ‘Toddler/Baby’
* 3 to 17 = ‘Child’
* 18 to 65 = ‘Adult’
* 66 to 99=’Elderly’


```python
# Creating Age
rng=np.random.RandomState(42)
age = pd.DataFrame({'age': rng.randint(1, 100, 20)})
age['bins'] = pd.cut(x=age['age'], bins=[0,2,17,65,99],
                    labels=['Baby','Child','Adult','Elderly'])
age.T
```

## **B. Groupby Operations**
In order to split the data, we use groupby() function this function is used to split the data into groups based on some criteria. Pandas objects can be split on any of their axes.

By **“group by”** we are referring to a process involving one or more of the following steps:

* Splitting the data into groups based on some criteria.

* Applying a function to each group independently.

* Combining the results into a data structure.

```python
df = pd.DataFrame([("bird", "Falconiformes", 389.0),
                   ("bird", "Psittaciformes", 24.0),
                   ("mammal", "Carnivora", 80.2),
                   ("mammal", "Primates", ' '),
                   ("mammal", "Carnivora", 58),],
                  index=["falcon", "parrot", "lion", "monkey", "leopard"],
                  columns=("class", "order", "max_speed"))
df
```

### **1. Aggregation**
Compute a summary statistic (or statistics) for each group. Some examples:
* Compute group sums or means.
* Compute group sizes / counts

```python
# Aggregation
summing = df.groupby(['class'])['max_speed'].sum()    # Berdasarkan kolom 'class', sum item 'max_speed'
meaning = df.groupby(['class'])['max_speed'].mean()   # Berdasarkan kolom 'class', average item 'max_speed'
counting = df.groupby(['class'])['max_speed'].count() # Berdasarkan kolom 'class', hitung jumlah item 'max_speed'

print(summing); print()
print(meaning); print()
print(counting)
```

### **2. Transformation**
Perform some group-specific computations and return a like-indexed object. Some examples:
* Standardize data (zscore) within a group.
* Filling NAs within groups with a value derived from each group.

```python
# Transformation
df.groupby(['class'])['max_speed'].transform(lambda x:x.fillna(x.mean())) # Mengisi data NA dengan mean data max_speed
```

### **3. Filtration** 
Discard some groups, according to a group-wise computation that evaluates True or False. Some examples:
* Discard data that belongs to groups with only a few members.
* Filter out data based on the group sum or mean.

```python
# Filtering
df.groupby(['class'])['max_speed'].filter(lambda x: x.mean()>200) # Filter item berdasarkan nilai mean 'max_speed' dari kelompok 'class'-nya 
```

```python
# Getting Groups
groups=df.groupby(['class'])
groups.get_group('bird')
```

```python
# Iterating through groups
for name,data in groups:
    print("\nName:",name)
    print("\n",data)
```
