# **Object Oriented Programming**

```python
class Furniture :
  def __init__ (self, name, material):
    self.name = name
    self.material = material
  def price (self, price):
    return f'The {self.material} {self.name} price is {price}'
  def fur_info(self):
    return f'I buy {self.material} {self.name}'

interior = Furniture('Table', 'Alumunium')

print(interior.fur_info())                                        # Instance Methods = price dipanggil dalam object interior
print(interior.price(250000))                                     
```

```python
class Furniture :
  def __init__ (self, name='Chair', material='Wood'):             # Default Object value
    self.name = name
    self.material = material
    self.colors = []

  def price (self, price):
    return f'The {self.material} {self.name} price is {price}'

  def fur_info(self):
    return f'I buy {self.material} {self.name}'

  def add_colors (self, color):
    self.colors.append(color)

interior1 = Furniture()
interior2 = Furniture('Table', 'Alumunium')
interior1.add_colors('Black')
interior1.add_colors('Brown')
interior1.add_colors('Dark Blue')

print(interior1.fur_info())                                        # Instance Methods = price dipanggil dalam object interior
print(interior2.price(250000)) 
print(interior1.colors)
```

# Encapsulated

```python
# Encapsulated 1

class Computer:

    def __init__(self):
        self.__maxprice = 900                                   # Private attribut

    def sell(self):
        print("Selling Price: {}".format(self.__maxprice))

    def setMaxPrice(self, price):
        self.__maxprice = price

c = Computer()
c.sell()

# ubah harga langsung
c.__maxprice = 1000
c.sell()

# ubah harga melalui fungsi
c.setMaxPrice(1000)
c.sell()
```

```python
class Mobil:
  def __init__(self, merk):
    self._merk = merk

class MobilBalap(Mobil):
  def __init__(self, merk, total_gear):
    super().__init__(merk)                                              # ambil atribut dari superclassnya (Mobil) yaitu merk
    self._total_gear = total_gear                                       # di set menjadi protect atribut
    self.__harga = 2000

  def ganti_harga(self, harga):
    self.__harga = harga

  def pamer(self):
    # akses _merk dari subclass
    print(
      f'Ini mobil {self._merk} dengan total gear {self._total_gear} seharga {self.__harga}'
    )

ferrari = MobilBalap('Ferrari', 8)
ferrari.pamer()
ferrari.ganti_harga(6000000)
ferrari.pamer()
```

# Inheritance

```python
# Inheritance

class Person:
  def __init__(self, first_name, last_name, age):
    self.first_name = first_name
    self.last_name = last_name
    self.age = age

class Student(Person):
  def __init__(self, first_name, last_name, age, region, city):
    super().__init__(first_name, last_name, age)
    self.region = region
    self.city = city
    self.skills = []

  def person_info(self):
    return f'{self.first_name} {self.last_name} is {self.age} years old. He lives in {self.city}, {self.region}'
  
  def add_skill(self,skill):
    self.skills.append(skill)

s1 = Student('Arthur', 'Curry', 33, 'England', 'London')
s2 = Student('Emily', 'Carter', 28, 'England', 'Manchester')
print(s1.person_info())
s1.add_skill('HTML')
s1.add_skill('CSS')
s1.add_skill('JavaScript')
print(s1.skills)

print(s2.person_info())
s2.add_skill('Organizing')
s2.add_skill('Marketing')
s2.add_skill('Digital Marketing')
print(s2.skills)
```

```python
# parent class
class Bird:   
    def __init__(self):
        print("Bird is ready")

    def whoisThis(self):
        print("Bird")

    def swim(self):
        print("Swim faster")

# child class
class Penguin(Bird):
    def __init__(self):
        # call super() function
        super().__init__()
        print("Penguin is ready")

    def whoisThis(self):
        print("Penguin")

    def run(self):
        print("Run faster")

peggy = Penguin()
```

# Polymorphism

```python
# Polymorphism

class Parrot:
    def fly(self):
        print("Parrot can fly")
    
    def swim(self):
        print("Parrot can't swim")

class Penguin:
    def fly(self):
        print("Penguin can't fly")
    
    def swim(self):
        print("Penguin can swim")

# common interface
def flying_test(bird):
    bird.fly()

#instantiate objects
blu = Parrot()
peggy = Penguin()

# passing the object
flying_test(blu)
flying_test(peggy)
```
