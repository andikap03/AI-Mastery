```python
# Importing Data processing libraries
import numpy as np 
import pandas as pd 
pd.plotting.register_matplotlib_converters()

# Plotting Libraries
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# Reading Data
url ='https://raw.githubusercontent.com/bluedataconsulting/AIMasteryProgram/main/Lab_Exercises/Module3/iris_data.csv'
iris=pd.read_csv(url)
print(f'Data shape : {iris.shape}')
iris.head()
```

```python
# Confirming the number of records for each species
iris['class'].value_counts()
```

```python
# Finding null count
iris.info()
```

```python
# Descriptive analysis
iris.describe()
```

# Univariate Analysis

```python
# Distinct Species values
iris["class"].unique()
```

```python
# Dividing data
setosa = iris.loc[iris['class'] == "Iris-setosa"]
versicolor = iris.loc[iris['class'] == "Iris-versicolor"]
virginica = iris.loc[iris['class'] == "Iris-virginica"]

iris.columns
```

```python
# Setting up subplots
f,axes=plt.subplots(ncols=2,nrows=4,figsize=(10,10),sharex=True)
l=['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i in range(4):
 
 #Ploting the boxplot
 sns.boxplot(x = iris.loc[:,l[i]], ax=axes[i][0], color='powderblue')
 
 # Ploting the KDE
 sns.histplot(data=iris.loc[:,l[i]], color="red", kde=True, stat="density", linewidth=0, ax=axes[i][1]) 
 plt.tight_layout()
```

# Bivariate Analysis

```python
# Set up the matplotlib figure
f, axes = plt.subplots(ncols = 2,figsize=(10, 4))
sns.despine(left=True)

# Sepal_length vs class
# Ploting the histogram with KDE
sns.histplot(data=setosa["sepal_length"],label="Iris-setosa",color='b',kde=True,linewidth=2,alpha=0.3, ax=axes[0])
sns.histplot(data=versicolor["sepal_length"],label="Iris-versicolor",kde=True,color='orange',linewidth=2,alpha=0.3, ax=axes[0])
sns.histplot(data=virginica["sepal_length"],label="Iris-virginica",kde=True,color='g',linewidth=2,alpha=0.3, ax=axes[0])

# Sepal_width vs class
# Ploting the histogram with KDE
sns.histplot(data=setosa["sepal_width"],label="Iris-setosa",color='b',kde=True,linewidth=2,alpha=0.3, ax=axes[1])
sns.histplot(data=versicolor["sepal_width"],label="Iris-versicolor",kde=True,color='orange',linewidth=2,alpha=0.3, ax=axes[1])
sns.histplot(data=virginica["sepal_width"],label="Iris-virginica",kde=True,color='g',linewidth=2,alpha=0.3, ax=axes[1])

plt.legend(loc='upper right')
plt.tight_layout()
```

# Multivariate Analysis

```python
#All species vs All Species
sns.pairplot(iris, hue="class")
plt.show()
```

```python
# Plotting Swarm and Box Plots
# Set up the matplotlib figure
f, axes = plt.subplots(ncols = 2, figsize=(8, 4), sharex=True)
sns.despine(left=True)

# Plot the Swarmplot
sns.swarmplot(x=iris['class'], y=iris['sepal_length'], ax=axes[0])

# Plot the Boxplot
sns.boxplot(x=iris['class'], y=iris['sepal_length'], ax=axes[1])

plt.tight_layout()
```

**Conclusion:**

Berdasarkan perbedaan plot diatas, sepal_length terlihat menarik untuk membedakan Class. Dapat disimpulkan bahwa :
- Sepal length dari Iris-setosa tergolong lebih kecil dari sepal length dua species yang lain.
- Meskipun perbedaan sepal length antara Iris-versicolor dan Iris-virginica tidak terlalu terlihat, sepal length dari Iris-versicolor tergolong lebih kecil dari sepal length Iris-virginica.
- Pada spesies Iris-Virginica terdapat nilai sepal length yang jauh berbeda dari nilai normalnya(terdapat outlier). 

```python
# Set up the matplotlib figure
f, axes = plt.subplots(ncols = 2, figsize=(8, 4), sharex=True)
sns.despine(left=True)

# Plot the Swarmplot
sns.swarmplot(x=iris['class'], y=iris['sepal_width'], ax=axes[0])

# Plot the Boxplot
sns.boxplot(x=iris['class'], y=iris['sepal_width'], ax=axes[1])

plt.tight_layout()
```

**Conclusion:**

Berdasarkan perbedaan plot diatas, sepal_length terlihat menarik untuk membedakan Class. Dapat disimpulkan bahwa :

- Sepal width dari Iris-virginica seems tergolong lebih lebar dari sepal width Iris-versicolor tapi untuk banyak sampel perbedaan ukuran ini tidak benar.
- Pada spesies Iris-Virginica terdapat nilai sepal width yang jauh berbeda dari nilai normalnya(terdapat outlier).

```python
# Set up the matplotlib figure
f, axes = plt.subplots(ncols = 2, figsize=(8, 4), sharex=True)
sns.despine(left=True)

# Plot the Swarmplot
sns.swarmplot(x=iris['class'], y=iris['petal_length'], ax=axes[0] ,size=4.5)

# Plot the Boxplot
sns.boxplot(x=iris['class'], y=iris['petal_length'], ax=axes[1])
plt.tight_layout()
```

**Conclusion:**

Berdasarkan perbedaan plot diatas, petal_length terlihat menarik untuk membedakan Class. Dapat disimpulkan bahwa :
* Petal length dari Iris-setosa tergolong lebih pendek daripada petal length dua species yang lain.
* Meskipun perbedaan petal length tidak erlalu jelas antara Iris-versicolor dan Iris-virginica, petal length dari Iris-versicolor tergolong lebih pendek daripada petal length dari Iris-virginica
* Pada spesies Iris-setosa dan Iris-versicolor terdapat nilai petal length yang jauh berbeda dari nilai normalnya(terdapat outlier).
