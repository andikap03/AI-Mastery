**5. List**
* Membuat List
* List Slicing dan Indexing
* Metode Basic List
* Nesting List

```python
# List
my_list = ['one','two', 4, 5.4]                                      # anggota list dapat dari berbagai tipe data

my_list.append('hai')                                                # menambah anggota list di akhir
print('Append method = ',my_list)

my_list.insert(2, 'three')                                           # menyisipkan ke posisi tertentu
print('Insert method = ',my_list)

my_list.extend([1, 2, 3])                                            # menambahkan anggota list lain ke bagian akhir list utama
print('Extend method = ',my_list)

my_list.remove('one')                                                # menghilangkan anggota list dengan menyebutkan namanya
print('Remove method = ',my_list)

my_list.pop(3)                                                       # menghilangkan anggota list dengan menyebutkan indexnya
print('Pop method = ',my_list)

# List Slicing
print(my_list[1:-1:2])
print(my_list[1::2])                                                 # kosong artinya keseluruhan (dr awal atau hingga akhir)
print(my_list[:-2:2])

# Nesting List
bucket = [['Bananas', 'Apple', 'Grape'], 'Meat', 'Vegetable']         
fruit = bucket[0][:3]                                                # [anggota section primer][anggota section sekunder][dst.]
print('Fruit types : ', fruit)
```

```python
# Membuat dictionary
my_dict = {'key1':'value1','key2':'value2'}
print(my_dict['key2'])                        # memanggil value dari key

my_dict['key1']=123                           # merubah nilai key
print(my_dict)

my_dict['key1'] = my_dict['key1'] - 123
print(my_dict['key1'])
 
my_dict['key1'] -= 123                        # -= adalah menghasilkan result negatif
print(my_dict['key1'])

# Membuat dictionary kosong baru
d = {}
d['animal'] = 'Dog'                           # mengisikan dictionary
print(d)

d['fruit'] = 'Apple'
print(d)
```

**7. Set**

```python
#Set
x = set()

x.add(1)                                  # menambahkan elemen set dengan metode add()
print(x)

x.add(2)                                  # menambahkan elemen baru
print(x)

x.add(1)                                  # menambahkan elemen yang sama = tidak bertambah
print(x)

l = [1,1,2,2,3,4,5,6,1,1]                 # membuat list dengan nilai acak
print(set(l))                             # mendapatkan nilai unik (tdk ada duplikat)

# Operasi dengan set
my_set = {1,2,3,4,5,1,2,3,7,8,4,9,0}
print(my_set)

my_set.add(10)                            # menambahkan 1 karakter ke anggota himpunan
print(my_set)

my_set.update([4,6,8,11])                 # menambahkan beberapa karakter ke anggota himpunan  
print(my_set)

my_set.remove(6)                          # menghilangkan anggota himpunan, jika anggota yg dimaksud tdk ada maka akan eror
print(my_set)

my_set.discard(12)                        # menghilangkan anggota himpunan, jika anggota yg dimaksud tdk ada maka akan tdk eror
print(my_set)
print()

# Operasi Himpunan
my_set1 = my_set
my_set2 = {2,4,15,1,12}

gabungan = my_set1.union(my_set2)                        # Gabungan berarti menggabung anggota 2 himpunan dan menghilangkan nilai yg sama
print('Himpunan Gabungan = ',gabungan)
print('Himpunan Gabungan = ', my_set1 | my_set2)

irisan = my_set1.intersection(my_set2)                   # Irisan berarti mencari anggota yang bernilai sama dari 2 himpunan
print('Himpunan Irisan = ', irisan)
print('Himpunan Irisan = ', my_set1 & my_set2)

selisih = my_set1.difference(my_set2)                    # Selisih A dan B berarti mencari anggota A yang tidak menjadi anggota B, dan sebaliknya
print('Himpunan Selisih = ', selisih)
print('Himpunan Selisih = ', my_set1 - my_set2)

komplemen = my_set1.symmetric_difference(my_set2)        # Komplemen berarti menggabungkan anggota A yg bukan anggota B dan anggota B yg bukan anggota A
print('Himpunan Komplemen = ', komplemen)
print('Himpunan Komplemen = ', my_set1 ^ my_set2)
```

**9. Looping**
* For Loop
* While Loop
* Break Looping
* Continue Looping
* Pass Loop

```python
# For Loop
for i in range(0, 5):
  print(i)
print()

# While Loop
j = 0
while j < 5:
  j += 1
  print(j)
print()

# Break Loop
j = 0
while j < 10:
  if j == 6:
    break
  j += 1
  print(j)
print()

# Continue Loop
for i in range(0, 10):
  if i%2 == 0:
    continue
  print(i)
print()

# Pass Loop
for i in range(0, 5):
  pass
print("Haiiiii")
```

**10. Comprehension**
* List Comprehension
* Dict Comprehension

```python
# List Comprehension

myList = [1,2,3,4,5,6]
kuadrat = [num**2 for num in myList]                                                                        # List Comprehension dengan for 
print(kuadrat)

kuadrat_genap = [num**2 for num in myList if num%2==0]                                                      # List Comprehension dengan for dan cabang if
print(kuadrat_genap)

kuadrat_kubik = [num**2 if num%2 == 0 else num**3 for num in myList]                                        # List Comprehension dengan for dan cabang if, else
print(kuadrat_kubik)

kuadrat_kubik_p4 = [nums**2 if nums%2 == 0 else nums**3 if nums%3 == 0 else nums**4 for nums in myList]     # List Comprehension dengan for dan cabang if, elif, else
print(kuadrat_kubik_p4)

# Dict Comprehension

myDict = {'apple': 20, 'banana': 30, 'grape': 40, 'kiwi': 50, 'melon': 60}
naikkan = {key:(value+10) for (key, value) in myDict.items()}                                                           # Dict Comprehension dengan for
print(naikkan)

diskon = {key:(value-10) for (key, value) in myDict.items() if value>40}                                                # Dict Comprehension dengan for dan if
print(diskon)

harga_baru = {key:(value-10 if value>40 else value) for (key, value) in myDict.items()}                                 # Dict Comprehension dengan for dan if else
print(harga_baru)

harga_baru2 = {key:(value-10 if value>40 else value*2 if value%3 == 0  else value) for (key, value) in myDict.items()}  # Dict Comprehension dengan for dan if, elif, else
print(harga_baru2)
```
